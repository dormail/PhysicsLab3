```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lif = pd.read_csv('data/lif_r1.csv')
NaCl1 = pd.read_csv('data/nacl_r1.csv')
NaCl2 = pd.read_csv('data/nacl_r2.csv')
ti = pd.read_csv('data/ti_r1.csv')
```

```python
plt.clf()

df = lif
name = 'LiF'

# finding max intensity
glance_ind = np.argmax(df['I'])
glance = df['theta'][glance_ind]
print(f'Glancing angle {name}: {glance}')


plt.scatter(df['theta'], df['I'], marker='+',
           label=f'Reflection {name}')

plt.plot([glance, glance], [0, np.max(df['I'])], c='r', label=rf'Glance reflection $\theta = {glance}^\circ$')

plt.xlabel(r'$\theta$')
plt.ylabel(r'Intensity / $(s^{-1})$')

plt.legend()

plt.tight_layout()
plt.savefig('build/LiF.pdf')
glance_lif = glance
```

```python
plt.clf()

ind0 = 80
ind1 = 150

df = NaCl1
name = 'NaCl'

# finding max intensity
data = df['I'][ind0:ind1]
glance_ind = np.argmax(data) + ind0
glance_2 = df['theta'][glance_ind]

plt.scatter(df['theta'][ind0:ind1], df['I'][ind0:ind1], marker='+',
           label=f'Reflection {name}')


plt.xlabel(r'$\theta$')
plt.ylabel(r'Intensity / $(s^{-1})$')

plt.legend()

plt.tight_layout()
```

```python
glance_2_rad = glance_2 * 2 * np.pi / 360
pm = 10**-12
d = 282.01 * pm

wavelength = 2 * d * np.sin(glance_2_rad) /2 
print(f'Wavelength = {wavelength / pm} pm')
```

```python
# literature values
pm = 10**-12

wavelength_alpha = 71.08 * pm
wavelength_beta = 63.09 * pm
```

```python
print('Ti lattice constants')
d = wavelength_alpha / (2 * np.sin(glance_ti * np.pi * 2 / 360))
a0 = 2 * d
print(f'assuming K_alpha a0 = {a0 / pm} pm')

d = wavelength_beta / (2 * np.sin(glance_ti * np.pi * 2 / 360))
a0 = 2 * d
print(f'assuming K_beta a0 = {a0 / pm} pm')
```
